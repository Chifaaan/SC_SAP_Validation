```python
kolom_tanggal_konsi_a = 'liph_date'
kolom_tanggal_konsi_b = 'posting_date'


konsi_b[kolom_tanggal_konsi_b] = pd.to_datetime(konsi_b[kolom_tanggal_konsi_b], errors='coerce')
konsi_a[kolom_tanggal_konsi_a] = pd.to_datetime(konsi_a[kolom_tanggal_konsi_a], errors='coerce')
```

```python
konsi_b = konsi_b.rename(columns={
    'profit_center': 'kode_outlet'})
```

```python
kolom_kunci = 'kode_outlet'

df_gabungan = pd.merge(
    konsi_b,
    konsi_a,
    on=kolom_kunci,
    how='outer',
    suffixes=('_sap', '_sc')
)

print("\nContoh data setelah digabungkan:")
print(df_gabungan.head())
```

```python
conditions = [
    (df_gabungan[kolom_tanggal_konsi_a] == df_gabungan[kolom_tanggal_konsi_b]), # Kondisi 1: Tanggal sama
    (df_gabungan[kolom_tanggal_konsi_a].isnull()) | (df_gabungan[kolom_tanggal_konsi_b].isnull()), # Kondisi 2: Salah satu data tidak ada
]
choices = [
    'Sesuai',           # Hasil jika Kondisi 1 benar
    'Data Tidak Lengkap' # Hasil jika Kondisi 2 benar
]

df_gabungan['Status_Validasi'] = np.select(conditions, choices, default='Tidak Sesuai')

print("\nData dengan Status Validasi:")
print(df_gabungan.head())
```
